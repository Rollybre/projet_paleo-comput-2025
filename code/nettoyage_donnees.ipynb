```python
import os

DOSSIER_RACINE = "/Users/rolly/Documents/10-19_Université_et_scolarité/paleo_comput/data_clean"  # ← À modifier
CARACTERES_INTERDITS = set('<>:\'"/\\|?*[];')    # ← À adapter à tes besoins
def nettoyer_nom(nom):
    return ''.join(c for c in nom if c not in CARACTERES_INTERDITS)

# Parcours récursif
for dossier, _, fichiers in os.walk(DOSSIER_RACINE):
    for ancien_nom in fichiers:
        nouveau_nom = nettoyer_nom(ancien_nom)

        if ancien_nom != nouveau_nom:
            chemin_ancien = os.path.join(dossier, ancien_nom)
            chemin_nouveau = os.path.join(dossier, nouveau_nom)

            # Gestion de conflits de noms (éviter d'écraser un fichier)
            if not os.path.exists(chemin_nouveau):
                os.rename(chemin_ancien, chemin_nouveau)
                print(f"✅ Renommé : {ancien_nom} → {nouveau_nom}")
            else:
                print(f"⚠️ Conflit : {nouveau_nom} existe déjà dans {dossier}")
```
